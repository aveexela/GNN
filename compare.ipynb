```python
import pandas as pd
from bert_score import score
import torch
```

```python
baseline_df = pd.read_excel("/kaggle/input/rag-answer-compilation-dataset/QA_dataset_answered_baseline.xlsx")
graph_df = pd.read_excel("/kaggle/input/rag-answer-compilation-dataset/QA_dataset_answered_v2.xlsx")

assert baseline_df["Question"].equals(graph_df["Question"]), "Вопросы не совпадают"
```

```python
# Список вопросов и ответов
questions = baseline_df["Question"].tolist()
baseline_answers = baseline_df["Answer"].astype(str).tolist()
graph_answers = graph_df["Answer"].astype(str).tolist()

# Сравнение по BERTScore
P, R, F1 = score(graph_answers, baseline_answers, lang="ru", verbose=True, device="cuda" if torch.cuda.is_available() else "cpu")

# Результаты
compare_df = pd.DataFrame({
    "Question": questions,
    "Baseline_Answer": baseline_answers,
    "Graph_Answer": graph_answers,
    "BERT_P": P.numpy(),
    "BERT_R": R.numpy(),
    "BERT_F1": F1.numpy()
})

# Средний результат
mean_f1 = compare_df["BERT_F1"].mean()

compare_df.to_excel("QA_dataset_comparison_v1_vs_v2.xlsx", index=False)
print(f"Средний BERTScore Graph-RAG vs Baseline: {mean_f1:.4f}")
compare_df.head()
```
